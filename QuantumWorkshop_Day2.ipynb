# Hands-On: The Knapsack Problem
(Gonciulea & Stefanski, 2025, p. 20)

The knapsack problem is a classic example of a binary optimization problem.
- Given a set of items, each with a weight and a value, and a knapsack with a maximum weight capacity, select the combination of items that maximizes total value without exceeding the weight limit.
- Each item represents a binary decision: include it (1) or exclude it (0).

- In this example we solve a small knapsack problem with a maximum weight capacity of 4 units and three items weighing 3, 2, and 1 units.
- Their values are \$2000, \$3000, and \$1000 (scaled down to 2, 3, and 1 for simplicity).
- The goal is to evaluate all possible item combinations and determine which combination maximizes value without exceeding the weight limit.

## Quantum Registers
- Similar to how classical programs use registers or to store values, **quantum registers** are groups of qubits used to represent different variables in a computation.
- In the knapsack problem, separate registers are used to track item selection, total value, and total weight.

- Three registers encoding selection, value, and weight
  <img src="https://raw.githubusercontent.com/FSCJ-FacultyDev/WC26-Summer-QuantumWorkshop/main/images/knapsack-figure2.4.png"
      alt="Three registers for Knapsack Problem"
      width="75%">

  - The number of qubits in each register depends on how much information must be represented for the problem.
  - The selection register uses one qubit per item to represent whether each item is included (1) or not (0).
  - Additional registers store the total value and weight, which are calculated as functions of the binary selection variables.
  - This structured encoding prepares the problem for implementation within a quantum circuit.

## Implementing a Quantum Knapsack Solver with Grover's Search
The knapsack problem can be solved using a solve_knapsack function that takes item values, weights, and a maximum weight limit as inputs.
- Algorithm summary: Initialize all possible item selections in superposition, encode each selection's total value and weight into quantum registers, and use Grover's search to amplify selections that satisfy the weight constraint and exceed the current value threshold (the minimum total value a solution must have to be considered a candidate). After each successful search, increase the minimum value threshold and repeat until no better solution is found.
- The implementation combines two key quantum techniques: function encoding, which computes the value and weight of each selection into quantum registers, and Grover's search, which uses an oracle and amplitude amplification to increase the probability of measuring valid, high-value solutions.
- The final measurement returns the optimal selection that maximizes value without exceeding the weight limit.

````python
from math import floor, log10, atan2, log2, pi

from sty import RgbBg, Style, bg, fg

from hume.utils.common import padded_bin
from hume.utils.common import complex_to_rgb

# Select the simulation backend: 'hume' uses the Hume educational quantum
# simulator, while the commented line can be enabled to use the Qiskit backend.
# The Hume simulator is designed specifically for learning and visualization,
# rather than performance or hardware compatibility. It exposes the internal
# quantum state directly, allowing users to see how amplitudes change during
# a computation.
SIMULATOR = 'hume'
# SIMULATOR = 'qiskit'

# define a function that formats and displays a quantum knapsack state as a
# table of decoded register values, amplitudes, phases, and amplitude bars.
def knapsack_table_new(
    state,
    key_size,
    value_size,
    weight_size,
    decimals=4,
    symbol='\u2588',
    colors=None
):
    n = int(log2(len(state)))
    round_state = [
        round(state[k].real, decimals)
        + 1j * round(state[k].imag, decimals)
        for k in range(len(state))
    ]

    headers = [
        'Selection',
        'Value',
        'Binary',
        'Weight',
        'Binary',
        'Magnitude',
        'Direction',
        'Amplitude Bar'
    ]
    offsets = [
        # max(len(headers[0]), floor(log10(len(state)))),  # outcome key
        max(len(headers[0]), key_size),                   # binary key
        max(len(headers[1]), floor(log10(len(state)))),   # outcome value
        max(len(headers[2]), value_size),                 # binary value
        max(len(headers[3]), floor(log10(len(state)))),   # outcome weight
        max(len(headers[4]), n - key_size - value_size),  # binary weight
        # max(len(headers[4]), 2 * (decimals + 2) + 6),   # amplitude
        max(len(headers[5]), decimals + 2),               # magnitude
        max(len(headers[6]), decimals),                   # direction
        max(len(headers[7]), 24),                         # amplitude bar
        # max(len(headers[8]), decimals + 2),             # probability
    ]

    for i in range(len(offsets)):
        headers[i] = headers[i] + ' ' * (offsets[i] - len(headers[i]))

    header_str = '  '.join(headers)

    output = '\n'
    # output += '\033[1m' + '      Key             Value' + '\033[0m'
    # output += '\n'
    output += header_str
    output += '\n'
    output += len(header_str) * '-'
    output += '\n'

    ind_to_color = []  # array containing the color for each row index

    # populate ind_to_color with None for each row
    for i in range(len(round_state)):
        ind_to_color.append(None)

    # add the color for each index that's supposed to be colored
    if colors is not None:
        for key in colors:
            for index in colors.get(key):
                # make sure each index is only assigned one color
                assert(not ind_to_color[index])
                ind_to_color[index] = key

    rows = []
    for k in range(len(round_state)):
        if abs(round_state[k]) > 0.0001:
            direction = round(
                atan2(round_state[k].imag, round_state[k].real) * 180 / pi,
                2
            )

            # if creating a row that has a background color, get the color
            if ind_to_color[k]:
                bg.my_color = Style(
                    RgbBg(
                        ind_to_color[k][0],
                        ind_to_color[k][1],
                        ind_to_color[k][2]
                    )
                )

            rows += [[
                padded_bin(n, k)[-key_size:].ljust(offsets[0], ' '),

                str(
                    int(padded_bin(n, k)[weight_size:-key_size], 2)
                ).ljust(offsets[1], ' '),

                padded_bin(n, k)[weight_size:-key_size].ljust(
                    offsets[2],
                    ' '
                ),

                str(
                    int(padded_bin(n, k)[0:weight_size], 2)
                ).ljust(offsets[3], ' '),

                padded_bin(n, k)[0:weight_size].ljust(offsets[4], ' '),

                str(round(abs(state[k]), decimals)).ljust(
                    decimals + 2,
                    ' '
                ).ljust(offsets[5], ' '),

                (
                    str(
                        (
                            (' ' if direction >= 0 else '-')
                            + str(floor(abs(direction)))
                        ).rjust(4, ' ')
                        + '.'
                        + str(
                            int(
                                100 * round(
                                    direction - floor(direction),
                                    2
                                )
                            )
                        ).ljust(2, '0')
                    ) + '\u00b0'
                    if abs(round_state[k]) > 0
                    else offsets[6] * ' '
                ).ljust(offsets[6], ' '),

                fg(*[int(255 * a) for a in complex_to_rgb(state[k])]) + (
                    int(abs(state[k] * 24)) * symbol
                ).ljust(offsets[7], ' ') + fg.rs

                # str(round(abs(state[k]) ** 2, decimals)).ljust(
                #     decimals + 2, ' '
                # ) + bg.rs
            ]]

    sorted_rows = sorted(rows, key=lambda x: int(x[0]))

    for row in sorted_rows:
        output += '  '.join(row)
        output += '\n'

    return output
````

````python
from math import ceil, log2

from hume import QuantumRegister, QuantumCircuit
from apps.utils import knapsack_table_new
from hume.algos.grover_optimizer import grover_optimizer, oracle_match_0_multi
from hume.algos.function_encoding import encode_term
from hume.utils.common import padded_bin

# Define and run a Grover-based quantum knapsack solver that builds the circuit,
# encodes item values and weights, searches for valid selections, and reports
# the best solution found.

def build_knapsack_circuit(n_key, n_value, n_weight, v, w):

    key = QuantumRegister(n_key)
    value = QuantumRegister(n_value)
    weight = QuantumRegister(n_weight)
    circuit = QuantumCircuit(key, value, weight)

    for i in range(len(key)):
        circuit.h(key[i])

    for i in range(len(value)):
        circuit.h(value[i])

    for i in range(len(weight)):
        circuit.h(weight[i])

    # encode value function
    for (coeff, vars) in v:
        encode_term(coeff, vars, circuit, key, value)

    circuit.iqft(value[::-1], swap=False)

    # encode weight function
    for (coeff, vars) in w:
        encode_term(coeff, vars, circuit, key, weight)

    circuit.iqft(weight[::-1], swap=False)

    return circuit


def terms(l):
    return  [(l[k], [k]) for k in range(len(l))]


def get_selection_value(selection_binary, v):
    val = 0
    for i in range(len(selection_binary)):
        val += int(selection_binary[::-1][i]) * v[i][0]

    return val


def get_selection_weight(selection_binary, w):
    weight = 0
    for i in range(len(selection_binary)):
        weight += int(selection_binary[::-1][i]) * w[i][0]

    return weight


def solve_knapsack(values, weights, max_weight):

    def action_on_progress(flow_state):
        n_key = flow_state['circuit_params']['n_key']
        n_value = flow_state['circuit_params']['n_value']
        n_weight = flow_state['circuit_params']['n_weight']
        s = flow_state['last_run_result']['state vector']
        print(knapsack_table_new(s, n_key, n_value, n_weight))

        print(f'\nLooking for values >= {
            flow_state["circuit_params"]["min_value"]}')


    assert(len(values) == len(weights))
    n_key = len(values)

    v = terms(values)
    w = terms(weights)

    v_adjusted = [(-max(values) - 0, [])] + v
    w_adjusted = [(-item[0], item[1]) for item in w] + [(max_weight, [])]

    min_value = max(values)
    n_value = ceil(log2(max(min_value, sum(values) - min_value))) + 1

    # weight register size based on w_adjusted
    n_weight = ceil(log2(sum([max_weight - item[0] for item in w]))) + 1

    oracle = oracle_match_0_multi(n_key + n_value + n_weight,
                                  [n_key + n_value - 1,
                                   n_key + n_value + n_weight - 1])

    def process_outcome(outcome, flow_state):
        n = flow_state['circuit_params']['n_key'] + \
              flow_state['circuit_params']['n_value'] + \
              flow_state['circuit_params']['n_weight']
        outcome_selection = padded_bin(n, outcome)[-n_key:]
        outcome_value = get_selection_value(outcome_selection, v)
        outcome_weight = get_selection_weight(outcome_selection, w)
        return outcome_selection, outcome_value, outcome_weight

    def update_circuit_params(outcome_results, flow_state):
        circuit_params = flow_state['circuit_params']
        outcome_selection, outcome_value, outcome_weight = outcome_results
        v = circuit_params['v']

        # adjust value function
        # v_adjusted = v + [(-outcome_value - 0, [])]
        v[0] = (-outcome_value - 1, [])

        # circuit_params['v'] = v_adjusted

        circuit_params['min_value'] = outcome_value + 1
        action_on_progress(flow_state)

    def progress(results, flow_state):
        outcome_selection, outcome_value, outcome_weight = results
        min_value = flow_state['circuit_params']['min_value']
        return (outcome_value >= min_value) and (outcome_weight <= max_weight)

    def build_circuit(flow_state):
        return build_knapsack_circuit(flow_state['circuit_params']['n_key'],
                                      flow_state['circuit_params']['n_value'],
                                      flow_state['circuit_params']['n_weight'],
                                      flow_state['circuit_params']['v'],
                                      flow_state['circuit_params']['w'])

    results = grover_optimizer({'n_key': n_key, 'n_value': n_value,
                                'n_weight': n_weight, 'v': v_adjusted,
                                'w': w_adjusted, 'min_value': min_value},
                     build_circuit, oracle,
                     update_circuit_params, progress, process_outcome,
                               failure_threshold=3)

    items_string = results[0]
    items = [str(len(items_string) - 1 - k)
      for k in range(len(items_string)) if items_string[k] == '1'][::-1]

    print(
      f'Optimal selection consists of items {", ".join(items)}. '
      f'The combined value is {results[1]} and the combined weight '
      f'is {results[2]}.'
    )

def test_knapsack():
    values = [2, 3, 1]
    weights = [3, 2, 1]
    max_weight = 4


    solve_knapsack(values, weights, max_weight)

test_knapsack()
````

## More on Single-Qubit States and Gates

A quantum state can be represented in code as a state vector consisting of a pair of amplitudes which describe the contribution of each basis state.
- As we saw in the previous workshop session, quantum states are often written as kets using Dirac notation:

  |ψ⟩ = α|0⟩ + β|1)

- e.g.,

  ∣ψ⟩=(-0.34+0.43i)|0⟩+(0.68+0.49i)|1⟩

- The qubit state shown above is described as a superposition of |0> and |1>, where the complex amplitudes determine the measurement probabilities after computing their squared magnitudes.
- For a complex amplitude a+bi, the probability is a^2 + b^2.

  |-0.34+0.43i|^2 = (-0.34)^2 + (0.43)^2 = 0.1156+0.1849 ≈ 0.3005

  |0.68+0.49i|^2 = (0.68)^2 + (0.49)^2 = 0.4624 + 0.2401 ≈ 0.7025

- Final probabilities
  - P(0) ≈ 0.3005
  - P(1) ≈ 0.7025

  - Note that these sum to approximately 1 (because the original amplitudes were rounded to two places, otherwise they would sum to exactly 1), which is required for a valid quantum state.

## Trigonometric Form of a Qubit State
A quantum state of a qubit can be expressed using trigonometric formulas by describing its amplitudes with angles.
- Instead of writing arbitrary complex numbers, the amplitudes are written using sine and cosine functions so that the probabilities automatically sum to 1.
- A common form (known as **Bloch sphere parameterization**) is

  |ψ> = cos(θ/2)|0> + e^(iφ) sin(θ/2)|1>

  or

  “psi equals cosine theta over two times ket zero, plus e to the i phi times sine of theta over two times ket one.”

  - (the constant e is used because it provides a convenient way to represent rotations and oscillations, fundamental to quantum mechanics).
  
  where the angles determine the magnitude and phase of the amplitudes.
- As we saw in the Day 1 workshop, these angles can be represented as a point on the Bloch sphere, allowing quantum states to be visualized as rotations on the sphere.
- We will mostly avoid this notation in this workshop; not because it isn’t important, but because saying things like “e to the i phi times sine of theta over two times ket one” before coffee can cause unnecessary stress.

## Hands-On: A Bloch Sphere Visualization Tool

The Bloch Sphere Visualizer at https://bloch.kherb.io/ is an interactive web tool for visualizing the state of a single qubit on a Bloch sphere.
- It provides an intuitive way to see how quantum states change when quantum gates or rotations are applied.
### Description
- Z axis (top and bottom): Represents the computational basis states |0⟩ and |1⟩, which determine the measurement outcomes.
- X axis (left to right across the equator): Represents equal superpositions of |0⟩ and |1⟩ with no phase difference, such as (|0⟩±|1⟩)/√2.
- Y axis (front to back across the equator): Represents equal superpositions where the amplitudes differ by an imaginary phase (±i), such as (|0⟩±i|1⟩)/√2.
### Demonstration
- Rotation Around the X Axis
  - When you rotate around the X axis, the state moves in a vertical circle that passes through the Y direction.
  - Think of it like spinning the sphere around the horizontal rod that goes left to right.
- Start State (INIT)
  - Start at |0⟩, which is the north pole.
- Apply an X-axis rotation of +90° (click +90° on the X axis)
  - The state moves from the north pole toward the front of the sphere, landing on the +Y axis.
    - Mathematically, this is approximately (|0⟩ + i|1⟩)/√2, a superposition with phase where the probabilities of measuring 0 or 1 are equal: P(0) = 0.5 and P(1) = 0.5.
    - This introduces an **i phase**, meaning the amplitude of |1⟩ is multiplied by i (i.e., √-1), creating a 90° phase difference between the amplitudes of |0⟩ and |1⟩.
    - This phase difference places the state on the +Y axis of the Bloch sphere.
    - This does not change measurement probabilities, but it affects how the state interferes with other states during computation.
- Rotate another +90° on the X axis (for a total 180°). Now the state reaches the south pole, or |1⟩
  - The 180° rotation around X flips the qubit (swaps the amplitudes of |0⟩ and |1⟩.
  - This is equivalent to the behavior of a quantum NOT gate (e.g., the **Pauli-X gate**), which flips the state of a qubit.
- Continue rotating another +90°
  - The state becomes (|0⟩ - i|1⟩)/√2, a superposition with equal probabilities but a -90° phase shift.
  - Now we are on the opposite side of the Y axis.
- Rotate a final +90° (360° total), returning to |0⟩.
- In general:
  - Rotation around the X axis mixes amplitudes and introduces phase (flips the qubit through superposition).
  - Rotation around the Y axis mixes amplitudes without introducing complex phase (moves the qubit smoothly between |0⟩ and |1⟩).
  - Rotation around the Z axis changes phase only (spins the phase around the equator while probabilities stay the same).
- Pauli Gates are basic single-qubit quantum gates that perform 180° rotations of a qubit’s state around the principal axes of the Bloch sphere.
  - Pauli-X (X): 180° rotation around the X axis (swaps |0⟩ and |1⟩, similar to a classical NOT gate)
  - Pauli-Y (Y): 180° rotation around the Y axis (swaps |0⟩ and |1⟩ while introducing phase)
  - Pauli-Z (Z): 180° rotation around the Z axis (changes the relative phase between |0⟩ and |1⟩ without changing measurement probabilities)

## Single Qubit with Example Amplitudes
- For a single qubit, there are two possible outcomes (0 and 1), each associated with a complex amplitude.
- These amplitudes determine the probabilities of measurement and can be represented in algebraic form within the table.
- Complex numbers can be visualized on the complex plane, where the horizontal axis represents the real part and the vertical axis represents the imaginary part.
- On this plane, the amplitude’s magnitude corresponds to its distance from the origin, and its direction corresponds to its phase (the angle from the real axis).
- The magnitude determines measurement probability (by squaring it, as we saw above), while the phase  influences how states interfere (combine and interact) during computation.
- The probabilities must sum to 1.
  - => if the probability of outcome 0 is p, then the probability of outcome 1 is 1 - p.
- These values can all be visualized in state tables using bars whose length represents magnitude and color represents direction.

### Polar Form
- A complex number can be represented by a magnitude r and an angle θ (which represents the direction).
- Using this information, we can write the number z = x + iy as
  - x = r cosθ
  - y = r sinθ
- which gives us
  - z = r(cos θ + i * sin θ)
- This is useful in quantum computing because quantum gates often rotate amplitudes by changing their phase θ while keeping their magnitude the same.
  - Many quantum operations are rotations, and rotations are much easier to describe with magnitude + angle than with real/imaginary parts.



## Programmatically Encoding Single-Qubit States with Python Lists
- A single-qubit quantum state can be expressed as a Python list of two amplitudes, where the index of an amplitude is the corresponding outcome.
- Python represents the imaginary unit, i, in a complex number with the symbol j.
  - j is used instead of i because in engineering i usually represents electric current; Python's creator (Guido van Rossum) used this convention as a design choice.
- To encode the state described above:
```
  state = [0.2958+0.51235j, -0.40311+0.69821j]
```

## More Basic Single-Qubit Gates
**Rx** (Rotation around X-axis)
- Rotates the qubit around the X-axis of the Bloch sphere by angle θ
- Mixes amplitudes (changes both magnitude and phase)
- Introduces complex (imaginary) components
- Useful for creating states with phase differences
- Can move a state out of the Z-axis into the Y direction

**Ry** (Rotation around Y-axis)
- Rotates the qubit around the Y-axis of the Bloch sphere by angle θ
- Adjusts amplitudes without introducing complex numbers
- Keeps amplitudes real-valued
- Acts like a "biased coin dial" between |0> and |1>

### Quick Contrast
- Rx: changes phase and probability (complex amplitudes).
  - Use it when you need phase manipulation for interference effects (key in quantum algorithms).
- Ry: changes probability only (real amplitudes).
  - Use it when you want intuitive control over measurement outcomes, e.g., applying Ry(π/2) to |0> creates an equal superposition.

## Single-Qubit Gate Inverses
- All quantum gates are reversible, meaning they have an inverse operation that can undo their effect.
- Some gates are self-inverse, meaning they undo themselves:
  - The X, Y, Z, and H gates all satisfy X² = I, Y² = I, Z² = I, and H² = I, meaning each gate is its own inverse.
  - Squaring (reapplying) these operations results in the **identity matrix** which means they "cancel themselves out" (the second application reverses the first).
```
      I =
        [1, 0]  
        [0, 1]
```
  - In the textbook you will see this represented using **U**, a general symbol for a quantum gate (unitary operator), allowing properties like inverses to be expressed in a standard, gate-independent way (e.g., U² = I).
- Other gates are reversed by applying the opposite operation:
  - For example, rotations:
    - Rz(θ)<sup>-1</sup> = Rz(-θ)
- Inverse notation is written using a power of -1:
  - e.g., X<sup>-1</sup> = X


## Example: Demonstrate that the X gate is its own inverse

- In the example shown below, the result of X @ X is:

```
    [1, 0]
    [0, 1]
```

- **np.allclose(X @ X, I)** checks whether two arrays are approximately equal, avoiding issues that arise when using == with floating-point values due to small rounding errors.
- It returns:
  - True → if all corresponding values are “close enough”
  - False → if they differ beyond a small tolerance
- The check returns True, which confirms that X² = I, meaning the X gate is self-inverse
- Note: The textbook represents qubit states as simple lists of amplitudes (e.g., [a, b]), while the example above uses a matrix representation (NumPy array) for gates and operations; both describe the same concepts; matrices can make it easier to demonstrate operations like gate multiplication. Just remember:
  - A qubit is a vector
  - A gate is a matrix
  - Applying a gate = matrix x vector

````python
import numpy as np

# Define the X gate
X = np.array([
    [0, 1],
    [1, 0]
], dtype=complex)

# Identity matrix
I = np.eye(2, dtype=complex)

# Show the gate
print("X =")
print(X)

# Verify X is its own inverse
print("\nX @ X =")
print(np.round(X @ X, 5))
print("Is X its own inverse?", np.allclose(X @ X, I))
````

## Transforming a Single Qubit State
- Now let's do some more experimentation with gate operations.
- First let's define functions to represent the gates
  - (but let's be good developers by including descriptive comments).



````python
from math import cos, sin, atan2, sqrt, pi
import numpy as np

# X gate
# Flips the qubit state: |0⟩ ↔ |1⟩
x = [
    [0, 1],
    [1, 0]
]

# Z gate
# Leaves |0⟩ unchanged and flips the phase of |1⟩
z = [
    [1, 0],
    [0, -1]
]

# Phase gate
# Applies a phase rotation to the |1⟩ amplitude by angle theta
def phase(theta):
    return [
        [1, 0],
        [0, complex(cos(theta), sin(theta))]
    ]

# Hadamard gate (H)
# Creates superposition: maps |0⟩ → (|0⟩ + |1⟩)/√2
h = [
    [1/sqrt(2),  1/sqrt(2)],
    [1/sqrt(2), -1/sqrt(2)]
]

# Rz rotation gate
# Rotates the qubit around the Z axis of the Bloch sphere by angle theta
def rz(theta):
    return [
        [complex(cos(theta / 2), -sin(theta / 2)), 0],
        [0, complex(cos(theta / 2),  sin(theta / 2))]
    ]

# Y gate
# Combines a bit flip with a phase change (uses imaginary values ±i)
y = [
    [0, complex(0, -1)],
    [complex(0, 1), 0]
]

# Rx rotation gate
# Rotates the qubit around the X axis by angle theta
def rx(theta):
    return [
        [cos(theta/2), complex(0, -sin(theta/2))],
        [complex(0, -sin(theta/2)), cos(theta/2)]
    ]

# Ry rotation gate
# Rotates the qubit around the Y axis by angle theta
def ry(theta):
    return [
        [cos(theta/2), -sin(theta/2)],
        [sin(theta/2),  cos(theta/2)]
    ]

# Textbook's transform function
# Apply a 2x2 quantum gate matrix to a single-qubit state vector,
# update amplitudes via matrix multiplication
#
# see below for a cleaner version using @
def transform_book(state, gate):
    assert(len(state) == 2)
    z0 = state[0]
    z1 = state[1]
    state[0] = gate[0][0]*z0 + gate[0][1]*z1
    state[1] = gate[1][0]*z0 + gate[1][1]*z1

# Cleaner version vs. book using numpy's @ operator
def transform(state, gate):
    assert len(state) == 2
    result = np.array(gate) @ np.array(state)
    state[0], state[1] = result


# Convert quantum state vector into table of amplitudes,
# phases in degrees, magnitudes, and probabilities
def to_table(s, decimals=5):
    table = [
        [
            k,
            s[k],
            atan2(s[k].imag, s[k].real) / (2 * pi) * 360,
            abs(s[k]),
            abs(s[k]) ** 2,
        ]
        for k in range(len(s))
    ]

    table_r = [
        [
            float(round(x, decimals))
            if isinstance(x, (float, np.floating))
            else complex(
                round(x.real, decimals),
                round(x.imag, decimals),
            )
            if isinstance(x, (complex, np.complexfloating))
            else x
            for x in row
        ]
        for row in table
    ]

    return table_r

# function to init qubit state to |0>
def init_state():
  # list comprehension to create list of state data
  state = [0 for _ in range(2)]
  # initialize amplitude (1 for outcome 0)
  state[0] = 1
  return state

def print_state(state, decimals=5):
  print(*to_table(state, decimals),sep='\n')

# Create a circuit by applying a sequence of gates
# to transform amplitudes and phase of qubit s

# initialize the state
s = init_state()
# apply gates in order
# Ry(2π/3) → X → P(π/3) → H
transform(s, ry(2*pi/3))
transform(s, x)
transform(s, phase(pi/3))
transform(s, h)
print_state(s)
````

## Real Hardware vs. Simulation

- Recall that quantum circuits are executed multiple times (shots) on real hardware
  - Each run ends with a measurement step
  - Measurement causes the quantum state to collapse to a single outcome
  - The observed result corresponds to the amplitude that becomes 1 after collapse
  - Individual runs may differ, but counts across many runs show the probability distribution

- Simulation, on the other hand, uses probabilities directly to mimic measurement outcomes
  - No need to rerun the full circuit each time; sampling replaces repeated execution
  - Outcomes are generated by randomly selecting one of the  measurement results based on probability weights (e.g., 0 or 1 for a single qubit)
  - This produces results that approximate the same distribution as real quantum hardware
  - This is faster and more efficient for small circuits, but it does not capture true quantum effects like noise or hardware constraint
  - The selection is weighted by their probabilities (e.g., 72% chance of 0, 28% chance of 1)
  - Each sample picks one outcome according to those weights, mimicking a real measurement

## Probabilistic Sampling
- We can use Python’s **random.choices** function to simulate quantum We can use Python’s random.choices function to simulate quantum measurement by sampling outcomes based on their probabilities
- Probabilities are computed as squared magnitudes of amplitudes:
```
    abs(s[k])**2
```
- This generates a list of measurement results (e.g., 10 samples → [0, 0, 0, 1, ...])
- **Counter** is used to count how often each outcome occurs, e.g., 0 -> 7, 1 -> 3
  - Counter is a Python class that counts elements in an iterable and returns their frequencies
- Small sample sizes approximate the distribution but can vary considerably
- Larger sample sizes (e.g., 1000) produce frequencies closer to true probabilities (e.g., 0 -> 726, 1 -> 274)
- This demonstrates how repeated measurements reveal the underlying probability distribution of a quantum state

````python
# simulate quantum measurement using probabilistic sampling

from random import choices
from collections import Counter

# 's' is a state vector of arbitrary amplitudes
s = [0.85+0j, 0.53+0j]

# Compute probabilities from amplitudes
probs = [abs(s[k])**2 for k in range(len(s))]

# --- Simulate 10 measurements ---
samples = choices(range(len(s)), probs, k=10)

print("Samples (10 shots):")
print(samples)

# Count outcomes
print("\nCounts (10 shots):")
for k, v in Counter(samples).items():
    print(f"{k} -> {v}")

# --- Simulate 1000 measurements ---
samples = choices(range(len(s)), probs, k=1000)

print("\nCounts (1000 shots):")
for k, v in Counter(samples).items():
    print(f"{k} -> {v}")
````

````python
from random import choices
from collections import Counter

# initialize the qubit in |0⟩
state = init_state()

# apply a gate that creates equal probabilities
transform(state, h)

# display the resulting state
print_state(state)

# compute probabilities from amplitudes
probs = [round(float(abs(state[k])**2), 5) for k in range(len(state))]

print(probs)

# simulate 10 measurements
samples = choices(range(len(state)), probs, k=10)

print("Samples:")
print(samples)

# count outcomes
print("\nCounts:")
for (k, v) in Counter(samples).items():
    print(str(k) + " -> " + str(v))

# simulate 1000 measurements
samples = choices(range(len(state)), probs, k=1000)

print("\nCounts (1000 shots):")
for (k, v) in Counter(samples).items():
    print(str(k) + " -> " + str(v))
````

### Reflection Questions
- How does the Hadamard gate create equal probabilities?
  - It transforms the qubit into a superposition with equal amplitudes, so squaring them produces equal probabilities (0.5 and 0.5).
- Why aren’t the counts exactly equal for 10 shots?
  - Measurement is random, so with a small sample size, the outcomes can vary due to chance.
- What happens as the number of shots increases?
  - The measured frequencies get closer to the true probabilities, approaching a 50/50 distribution.
